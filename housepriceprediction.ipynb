```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
from scipy import stats

from sklearn.feature_selection import mutual_info_regression
from sklearn.model_selection import train_test_split
from category_encoders import MEstimateEncoder

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# EDA and preprocessing
df = pd.read_csv('/content/data.csv')
df.head()
```

```python
df.info()
```

```python
# let's check the distribution of our target column
sns.kdeplot(df['price'].apply(np.log1p), fill=True)
plt.show()
```

```python
# From the data above we can see that the data is left skewed indicating that there maybe outliers
# Function to create Z-Score
def make_mi_score(x,y):
    x = x.copy()
    for colname in x.select_dtypes(['object','category']):
        x[colname], _ = x[colname].factorize()

    discrete_features = [pd.api.types.is_integer_dtype(t) for t in x.dtypes]
    mi_score = mutual_info_regression(x, y, discrete_features=discrete_features, random_state=42)
    mi_score = pd.Series(mi_score, name='Mutual Information Score', index=x.columns)
    mi_score = mi_score.sort_values(ascending=False)
    return mi_score
```

```python
X = df.drop(columns=['price'])
Y = df[['price']]

mi_score = make_mi_score(X, y)
mi_score
```

```python
# Function to plot the findings.
def plot_mi_score(score):
  score = score.sort_values(ascending = True)
  width = np.arange(len(score))
  ticks = list(score.index)
  plt.barh(width,score)
  plt.yticks(width,ticks)
  plt.title("Mutual Information Score")
plot_mi_score(mi_score)
```

```python
# Street and other information has a very high correlation with our target column price so lets redifine the dataset with the relevant informations
# Refining the dataset based on mutual information score
df = df[['street','statezip','city','sqft_living','sqft_above','bathrooms','yr_built','sqft_lot','bedrooms']]
df = df.join(y)
df
```

```python
# Using z-score to deal with outliers
z = np.abs(stats.zscore(df[['sqft_living','sqft_above','bathrooms','yr_built','sqft_lot','bedrooms']]))
z
```

```python
# A z-score of 3 or above indicates a significant deviation from the mean therefore making it an outlier
# Defining a threshold of 3
threshold = 3
print(np.where(z>3))
# The first array contains the list of row numbers and the second array respective column numbers, which have Z-score higher than 3.
# Let's remove the outliers and get the clean data!
df = df[(z<3).all(axis=1)]
df.info()
```

```python
# Training and testing the dataset
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
# Target Encoding
encoder = MEstimateEncoder(cols= ['street','statezip','city'],m=0.5)
x = encoder.fit_transform(x,y)
x_train,x_test,y_train,y_test = train_test_split(x.values,y.values,test_size = 0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score

cvs = cross_val_score(model, x_train, y_train, cv=10, n_jobs=-1)
print('Accuracy: {:.2f} %'.format(cvs.mean() * 100))
```

```python
y_pred = model.predict(x_test)
# Evaluating the model.
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print('mse',mse)
print('r2',r2)
```
